Import Libraries

```python
from matplotlib import pyplot as plt
import numpy as np
```

Simple Plot

```python
# Now add a title and labels to the plot

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))

ax.set_title("Simple Plot")
ax.set_xlabel(" X Label")
ax.set_ylabel("Y Label")
```

Multiple Lines

```python
#Now add Legends in the Graph

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x), label = 'sin', color = 'red')
ax.plot(x,np.cos(x),label = 'Cos', color = 'k')

ax.set_title('Multiple Lines')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

#ax.lagend()
ax.legend(loc = 1)
plt.show()
```

Line Style

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x), label = 'sin', color = 'green', linestyle = 'dashed')
ax.plot(x,np.cos(x),label = 'Cos', color = 'blue', linestyle = 'dotted')
ax.plot(x, np.sin(x+1), label = 'cos', linestyle = 'dashdot')

ax.set_title('Multiple Lines')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

#ax.lagend()
ax.legend(loc = 1)
plt.show()
```

Axes Limits


```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.set_xlim(-5, 15)
ax.set_ylim(-3, 3)
```

Multiple SubpPlots

```python
fig, axs = plt.subplots(2, 2, figsize=(10,6)) # a figure with a 2x2 grid of Axes
x = np.linspace(0, 10, 1000)

axs[0,0].plot(x, np.sin(x))

axs[0,1].plot(x, np.cos(x), color = 'maroon')

axs[1,0].plot(x, np.cos(x+2), color = 'blue')

axs[1,1].plot(x, np.sin(x+4), color = 'green');
```

```python
Labels =  ['G1','G2','G3','G4','G4']
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]

x = np.arange(len(Labels))
width = 0.25

fig, ax = plt.subplots()

ax.grid(linestyle='--', color='0.75', axis = 'y')

ax.set_axisbelow(True)    # Grid behind bars

rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(Labels)

ax.legend()

# Add bar labels
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

ax.set_ylim(0,40)
```

```python
fig, ax = plt.subplots()
ax.grid(linestyle='--', color='0.75', axis = 'y');
ax.set_axisbelow(True)

p1 = ax.bar(Labels, men_means, width, label = 'Men')
p2 = ax.bar(Labels, women_means, width, bottom = men_means, label = 'Women')

ax.set_title('Score by Group and Gender')
ax.set_xlabel('Group')
ax.set_ylabel('Scores')
ax.legend()

ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)
ax.set_ylim(0,70);
```

Scatter Plots

```python
x = np.random.randn(20)
y = np.random.randn(20)
```

 Add Markers

```python
fig, ax = plt.subplots()
ax.scatter(x,y,color = 'blue', marker = '*')
```

```python
fig, axs = plt.subplots(2, 3, sharex = True, sharey = True, figsize= (16,12))

#Marker Symbol
axs[0,0].scatter(x,y, s=80, marker = '>')
axs[0,0].set_title("marker='>'")

axs[0,1].scatter(x,y, s=80, marker = r'$\alpha$')
axs[0,1].set_title(r"marker=r'\$\alpha\$'")

#Marker from Path
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
axs[0,2].scatter(x,y, s=80, marker = verts)
axs[0,2].set_title("marker=verts")

#Regular Polygon Marker
axs[1,0].scatter(x,y, s=80, marker = (5,0))
axs[1,0].set_title("marker= Polygon")

#Regular Star Marker
axs[1,1].scatter(x,y, s=80, marker = (5,1))
axs[1,1].set_title("marker= Star")

# Regular Asterisk Marker
axs[1,2].scatter(x,y, s=80, marker = (5,2))
axs[1,2].set_title("marker= Asterisk")
```
